```python
# KAMUS
# Item, Kode, kode_bisa             : Array of strings
# Harga, uang_rupiah                : Array of integer
# cash_awal, harga_min, harga_maks  : Integer
# sisa_cash, saldo                  : Integer
# Price, count, count_eror          : Integer
# state_cash, state_kartu           : Boolean
# state_kartu1, state_cash1         : Boolean
# state1, state2                    : Boolean
# Kondisi                           : Boolean
# metode                            : String
# kode_transaksi, kode_pembelian    : String
# minuman                           : String
# masukan_uang()                    : Subprogram
# analisis_erorK(), analisis_erorK(): Subprogram


# Header dan Judul Program
print("\n===============================================================")
print("              Selamat Datang di Vending Machine")
print("                          Zeo Drinks")
print("===================== Selamat Berbelanja! =====================\n")

# Algoritma
import random
Item=['Air Mineral','Susu Pisang','Matcha','Susu Plain','Boba Milk','Yakult','Kopi Susu','Red Velvet','Jus Buah','Ocha'] # Data Item yang tersedia
Kode=['01','02','03','04','05','06','07','08','09','10'] # Kode Item
Harga=[1000, 6500, 7500, 6000, 8000, 2500, 5000, 9500, 5500, 4000] # Harga Item
uang_rupiah = [100, 200, 500, 1000, 2000, 5000, 10000, 20000, 50000, 100000]  # Pecahan uang rupiah
harga_min = 1000        # harga minuman termurah
harga_maks = 9500       # harga minuman termahal
cash_awal = 0           # inisialisasi
state_cash = False      # Inisialisasi
state_kartu = False     # Inisialisasi

# Menampilkan daftar minuman yang tersedia
print("          Daftar Item")
print("--------------------------------")
print("Minuman\t\tKode\tHarga")
print("--------------------------------")
for i in range (len(Item)):
    print (str(i+1), Item[i], "\t", Kode[i], "\tRp" , Harga[i] )
    i += 1
print("--------------------------------")

# Subprogram
def masukan_uang(uang_total):       # Menghitung total uang masukan user
    print("\nMasukan Uang Anda/ketik '0' jika selesai")
    kondisi = True                  # Inisialisasi
    while kondisi == True:
        # Meminta user memasukan nominal uang
        x = str(input(">>> Rp"))

        # Mengecek apakah yang diinput adalah nominal pecahan uang rupiah atau bukan
        try:
            uang_plus = int(x)
            if uang_plus in uang_rupiah:
                uang_total += uang_plus
            elif uang_plus == 0:
                kondisi = False
            else :
                print("Pastikan anda memasukan pecahan uang rupiah")
        except :
            print("Yang anda masukan bukan uang")
    return uang_total

def analisis_erorK(x):
    try:
        A = int(x)
        kata = "Kode minuman tidak ada/saldo anda kurang"
    except:
        kata = "Kode anda salah"
        if x == "":
            kata = "Anda tidak memasukan kode apapun"
    print(kata)

def analisis_erorC(x):
    try:
        A = int(x)
        kata = "Kode minuman tidak ada/uang anda kurang"
    except:
        kata = "Kode anda salah"
        if x == "":
            kata = "Anda tidak memasukan kode apapun"
    print(kata)

# Meminta user menginput kode metode pembayaran
print("\nMetode Pembayaran : ")
print(">> Kartu (K) \t: Jika pembayaran menggunakan kartu")
print(">> Cash (C) \t: Jika pembayaran menggunakan cash")
metode = str(input("Masukkan metode pembayaran (K/C): \n>>> "))

# Percabangan mengidentifikasi metode pembayaran yang diinginkan user
if metode == "K":

    print("Silakan memasukkan kartu anda\nKartu anda berhasil terbaca")
    saldo = random.randint(0,20000)         # Untuk mendapatkan angka sembarang sebagai saldo awal
    print("\n---------------------------------------------------------------")
    print("Petunjuk Memasukkan kode")
    print("1. Masukkan dua digit kode minuman yang sesuai jika anda \n   ingin membeli minuman")
    print("2. Ketik 'end' apabila anda ingin mengakhiri/membatalkan \n   pembelian")
    print("3. Transaksi otomatis dibatalkan/diakhiri anda memasukan \n   kode salah berulangkali atau jika saldo anda kurang")
    print("---------------------------------------------------------------")
    state_kartu = True

elif metode == "C":
    print("\n---------------------------------------------------------------")
    print("Petunjuk Memasukkan kode")
    print("1. Masukkan kode minuman yang sesuai jika anda ingin membeli \n   minuman")
    print("2. Ketik 'end' apabila anda ingin mengakhiri/membatalkan \n   pembelian")
    print("3. Ketik 'cash' jika anda ingin menambah uang")
    print("---------------------------------------------------------------\n")

    state_cash = True
    sisa_cash = cash_awal
    sisa_cash = masukan_uang(sisa_cash)     # Memanggil fungsi masukan_uang
    print(f"Total Uang = {sisa_cash}")

else :
    print("\nMetode pembayaran yang ada masukkan salah")

# Looping : proses pembelian menggunakan kartu
while state_kartu == True:
    # Mengecek saldo kartu user
    if saldo >= harga_maks :
        print(f"\nSaldo Anda adalah Rp{saldo}")
        print("\nMasukkan Kode")
        kode_transaksi = str(input("(dua digit kode minuman / 'end'):\n>>> ")) # Meminta user menginput kode minuman atau 'end' untuk berhenti

        if kode_transaksi in Kode:
            for i in range(len(Item)):
                if kode_transaksi == Kode[i] :
                    Price = int(Harga[i])   # Harga item yang dibeli
                    minuman = Item[i]       # Nama minuman yang dibeli
                    saldo -= Price
                i += 1
            print(f"\nSilakan mengambil '{minuman}' pesanan anda")
            print("---------------------------------------------------------------")

        elif kode_transaksi == "end":
            state_kartu = False             # Proses pembelian akan berakhir
        else :
            count = 1   # Inisialisasi
            analisis_erorK(kode_transaksi)

            # Looping input kode ulang apabila user menginput kode salah
            for i in range (2):
                print("\nMasukan kode ulang dan pastikan kode anda benar")
                kode_transaksi = str(input("(dua digit kode minuman / 'end'):\n>>> "))
                if kode_transaksi in Kode:
                    for i in range(len(Item)):
                        if kode_transaksi == Kode[i] :
                            Price = int(Harga[i])
                            minuman = Item[i]
                            saldo -= Price
                        i += 1
                    print(f"\nSilakan mengambil '{minuman}' pesanan anda")
                    print("---------------------------------------------------------------")
                    break
                elif kode_transaksi == "end":
                    state_kartu = False     # Proses pembelian akan berakhir
                    break
                else :
                    analisis_erorK(kode_transaksi)
                    count += 1
                i += 1
            if count == 3:
                print("\nMaaf, Anda telah memasukan kode salah 3x berturut-turut \nTransaksi diakhiri oleh sistem")
                state_kartu = False

    elif harga_min <= saldo < harga_maks :

        # Menampilkan saldo dan kode minuman yang dapat dibeli
        print(f"\nSaldo Anda adalah Rp{saldo}")
        print ("Kode Minuman yang dapat anda beli: \n>>", end=" ")
        kode_bisa = []      # Inisialisasi Array of string
        for i in range (len(Item)):
            if int(Harga[i]) <= saldo:
                print (Kode[i], end=" ")
                kode_bisa.append(Kode[i])
            i+=1

        print("\n\nMasukkan Kode")
        kode_transaksi = str(input("(dua digit kode minuman / 'end'):\n>>> ")) # Meminta user menginput kode minuman atau end untuk berhenti
        state_kartu1 = True # Inisialisasi
        # Looping
        while state_kartu1 == True :
            if kode_transaksi in kode_bisa:
                for i in range(len(Item)):
                    if kode_transaksi == Kode[i] :
                        Price = int(Harga[i])
                        minuman = Item[i]   # Harga item yang dibeli
                        saldo -= Price      # Nama minuman yang dibeli
                    i += 1
                print(f"\nSilakan mengambil '{minuman}' pesanan anda")
                print("---------------------------------------------------------------")
                break

            elif kode_transaksi == "end":
                state_kartu1 = False

            else :
                count = 1   # Inisialisasi
                analisis_erorK(kode_transaksi)

                # Looping input kode ulang apabila user menginput kode salah
                for i in range (2):
                    print("\nMasukan kode ulang dan pastikan kode anda benar")
                    kode_transaksi = str(input("(dua digit kode minuman / 'end'):\n>>> "))
                    if kode_transaksi in kode_bisa or kode_transaksi == "end":
                        break
                    else :
                        analisis_erorK(kode_transaksi)
                        count += 1
                    i += 1
                if count == 3:
                    print("\nMaaf, Anda telah memasukan kode salah 3x berturut-turut \nTransaksi diakhiri oleh sistem")
                    state_kartu1 = False

            if state_kartu1 == False:
                state_kartu = False     # Akan mengakhiri proses pembelian dan keluar dari looping

    elif saldo < harga_min :
        print(f"\n\nSaldo Anda adalah Rp{saldo}")
        print("Saldo tidak mencukupi untuk melanjutkan transaksi \nTransaksi diakhiri oleh sistem")
        state_kartu = False

# Looping : proses pembelian menggunakan cash
while state_cash == True:
    state_cash1 = True      # Inisialisasi
    state1 = False          # Inisialisasi
    state2 = True           # Inisialisasi
    kode_bisa = []          # Inisialisasi

    # Percabangan untuk mengecek sisa_cash
    if sisa_cash < harga_min:
        print("\nUang anda tidak cukup")
        print("Anda harus memasukan uang tambahan atau mengakhiri transaksi")
        print("Masukkan kode")
        Kode_pembelian = str(input("('cash'/'end'):\n>>> "))     # Meminta user input kode 'cash' untuk menambah uang / 'end' untuk berhenti
        # Percabangan untuk mengidentifikasi kode yang diinput user
        if Kode_pembelian == "cash":
            sisa_cash = masukan_uang(sisa_cash) # User memasukan uang tambahan
            print(f"Total Uang = {sisa_cash}")

        elif Kode_pembelian == "end":
            state_cash = False  # Proses pembelian akan berakhir
            break
        else :
            count_eror = 1
            for i in range (2):
                print("\nMasukan kode ulang dan pastikan kode anda benar")
                Kode_pembelian = str(input("('cash'/'end'):\n>>> "))   # Meminta user input kode ulang
                if Kode_pembelian == "cash":
                    sisa_cash = masukan_uang(sisa_cash) # User memasukan uang tambahan
                    print(f"Total Uang = {sisa_cash}")
                    break
                elif Kode_pembelian == "end":
                    state_cash1 = False
                    break
                else :
                    count_eror += 1
                i += 1
            if count_eror == 3:
                print("\nMaaf, Anda telah memasukan kode salah 3x berturut-turut \nTransaksi diakhiri oleh sistem")
                state_cash1 = False
                break
            state_cash = state_cash1    # Jika bernilai False maka program akan berhenti

    elif sisa_cash >= harga_min :
        state1 = True
        # Menampilkan kode minuman yang dapat dibeli
        print ("\nKode Minuman yang dapat anda beli:")
        print(end="")
        for i in range (len(Item)):
            if int(Harga[i]) <= sisa_cash:
                print (Kode[i], end=" ")
                kode_bisa.append(Kode[i])
            i+=1
        # Meminta user input kode minuman / 'cash' untuk menambah uang / 'end' untuk berhenti
        Kode_pembelian = str(input("\nMasukkan kode\n(dua digit kode minuman/'cash'/'end'):\n>>> "))

    while state1 == True and state2 == True :
        if Kode_pembelian in kode_bisa:
            for i in range(len(Item)):
                if Kode_pembelian == Kode[i] :
                    Price = int(Harga[i])       # Harga item yang dibeli
                    minuman = Item[i]           # Nama minuman yang dibeli
                    sisa_cash -= Price
            i += 1
            print(f"\nSilakan mengambil '{minuman}' pesanan anda")
            print("---------------------------------------------------------------")
            state2 = False
            break
        elif Kode_pembelian == "cash":
            sisa_cash = masukan_uang(sisa_cash) # Memanggil subprogram masukan_uang
            print(f"Total Uang = {sisa_cash}")
            break
        elif Kode_pembelian == "end":
            state_cash = False                  # Proses pembelian akan berakhir
            break
        elif Kode_pembelian =="" or Kode_pembelian != "cash" or Kode_pembelian != "end" or Kode_pembelian not in kode_bisa:
            analisis_erorC(Kode_pembelian)
            count_eror = 1
            for i in range (2):
                Kode_pembelian = str(input("\nMasukkan kode ulang dan pastikan kode anda benar\n(dua digit kode minuman/'cash'/'end'):\n>>> "))
                if Kode_pembelian in kode_bisa or Kode_pembelian == "cash" or Kode_pembelian == "end":
                    break
                else :
                    analisis_erorC(Kode_pembelian)
                count_eror += 1
                i+=1
            if count_eror == 3:
                print("\nMaaf, Anda telah memasukan kode salah 3x berturut-turut \nTransaksi diakhiri oleh sistem")
                state1 = False
            state_cash = state1     # Jika bernilai false maka akan keluar dari looping dan mengakhiri transaksi

if metode == "K":
    print("\n--------------------------")
    print("Silakan ambil kartu anda")
elif metode == "C":
    print("\n---------------------------------------------------------")
    print(f"Jangan lupa mengambil sisa uang anda, sisa uang : Rp{sisa_cash}")


print("===============================================================")
print("          Terima Kasih Telah Menggunakan Layanan Kami")
print("===============================================================")

```
